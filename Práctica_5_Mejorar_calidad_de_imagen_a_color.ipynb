```python
#Esdras García Lira 1995008
#America de los Ángeles Martínez Gámez  1933618

import cv2
import matplotlib.pyplot as plt
# Primero instalamos las bibliotecas para el procesamiento de imágenes y para
# la visualización. Este paso ya es algo rutinario asi que no entraremos en
# muchos detalles como la descripcion de cada biblioteca.

# Cargar imagen oscura
img_bgr = cv2.imread('/content/IMAGEN 1.jpg') #UTILIZAMOS LA IMAGEN PROPORCIONADA EN LA DESCRIPCION DEL TRABAJO
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Convertir a HSV
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

# Separar canales H, S y V
h, s, v = cv2.split(img_hsv)

# Ecualizar solo el canal de valor (brillo)
v_eq = cv2.equalizeHist(v)

# Combinar canales con el V ecualizado
img_hsv_eq = cv2.merge([h, s, v_eq])

# Convertir de nuevo a BGR y luego a RGB para visualización
img_eq_bgr = cv2.cvtColor(img_hsv_eq, cv2.COLOR_HSV2BGR)
img_eq_rgb = cv2.cvtColor(img_eq_bgr, cv2.COLOR_BGR2RGB)

# Mostrar imágenes original y mejorada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_eq_rgb)
plt.title('Imagen mejorada (Ecualización V)')
plt.axis('off')

plt.tight_layout()
plt.show()

```

En esta seccion del codigo primeramente como venimos haciendo de costumbre, instalamos las bibliotecas necesarias, no entraremos en detalles pues es algo que venimos haciendo todo el semestr, pero en general son para que el codigo funcione correctamente. Pasando de lleno a lo que el codigo realiza, primero se carga una imagen oscura en formato BGR y se convierte a RGB para su correcta visualización. Luego, se transforma al espacio de color HSV, donde se separan los canales de tono (H), saturación (S) y valor (V). Se aplica una ecualización de histograma únicamente al canal V para mejorar el brillo y el contraste de la imagen. Después, se vuelven a combinar los canales con el nuevo V ecualizado, se convierte la imagen de nuevo a RGB, y finalmente se muestra junto a la imagen original para comparar la mejora visual.

La imagen se reconstruye con el canal V mejorado y se convierte nuevamente a RGB, despues se muestra junto a la imagen original para resaltar el efecto de la ecualización en la visibilidad de los detalles.

```python
# Cargar imagen oscura
img_bgr = cv2.imread('/content/IMAGEN 3.jpg') # UTILIZAMOS UNA TERCERA IMAGEN MEDICA DE NUESTRA ELECCIÓN
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Convertir a HSV
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

# Separar canales H, S y V
h, s, v = cv2.split(img_hsv)

# Ecualizar solo el canal de valor (brillo)
v_eq = cv2.equalizeHist(v)

# Combinar canales con el V ecualizado
img_hsv_eq = cv2.merge([h, s, v_eq])

# Convertir de nuevo a BGR y luego a RGB para visualización
img_eq_bgr = cv2.cvtColor(img_hsv_eq, cv2.COLOR_HSV2BGR)
img_eq_rgb = cv2.cvtColor(img_eq_bgr, cv2.COLOR_BGR2RGB)

# Mostrar imágenes original y mejorada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_eq_rgb)
plt.title('Imagen mejorada (Ecualización V)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

Se procesa una tercera imagen médica oscura repitiendo el mismo procedimiento: carga en BGR, conversión a RGB, transformación a HSV, separación de canales y ecualización del canal V. Con esto se mejora el brillo de la imagen, que es especialmente útil para resaltar detalles clínicos o diagnósticos. Finalmente, se reconstruye la imagen, se convierte a RGB y se muestra comparativamente con la imagen original para visualizar la mejora obtenida.
